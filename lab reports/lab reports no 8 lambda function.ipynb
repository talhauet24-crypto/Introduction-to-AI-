### Task 1 — Basic lambda for a simple electrical formula
Write a lambda that takes a voltage value V and returns V² + 2.   
Test it with V = 5.    
(Represents a toy “adjusted voltage metric.”)   


```python
# Lambda function for adjusted voltage
adjusted_voltage = lambda V: V**2 + 2

# Test with V = 5
V = 5
result = adjusted_voltage(V)
print("Adjusted voltage for V =", V, "is", result)
```
