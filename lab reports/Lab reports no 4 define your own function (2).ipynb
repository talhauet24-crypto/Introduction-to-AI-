In circuit design, you often calculate series resistance.    
•	Write a function series() that takes a list or tuple of resistances.     
•	Return the total resistance.     
•	Add docstrings and annotations.     
•	Print the function’s __annotations        


```python
from typing import Iterable

def series(resistances: Iterable[float]) -> float:
    """
    Calculate the total resistance of resistors connected in series.

    Args:
        resistances (Iterable[float]): A list or tuple of resistance values (ohms).

    Returns:
        float: Total series resistance.
    """
    return sum(resistances)


# Example usage
total = series([10.0, 22.0, 47.0])
print("Total Resistance:", total)

# Print function annotations
print(series.__annotations__)
```
