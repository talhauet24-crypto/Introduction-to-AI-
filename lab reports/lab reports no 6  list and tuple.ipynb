### Lab Report Tasks:
Scenario:
You are designing a small student management system. You need to maintain student data for a class. Each student has a name, a roll number, and a list of grades. The system must allow grade updates but keep roll numbers fixed.
Tasks:
1.	Create a list of student names: ["Ali", "Usman", "Hassan", "Kamran"].
2.	Create a tuple of student roll numbers: (101, 102, 103, 104).
3.	For each student, create a list of grades (3 subjects) using nested lists:
Example: [ [85, 90, 78], [88, 92, 80], ... ]
4.	Update Ali’s second grade to 95.
5.	Add a new grade for Hassan in the fourth subject.
6.	Attempt to change Kamran’s roll number and observe what happens. Explain why.
7.	Print all students' names with their updated grades in a readable format.


```python
# Task 1: List of student names
students = ["Ali", "Usman", "Hassan", "Kamran"]

# Task 2: Tuple of roll numbers (fixed)
roll_numbers = (101, 102, 103, 104)

# Task 3: Nested list of grades (3 subjects each)
grades = [
    [85, 90, 78],  # Ali
    [88, 92, 80],  # Usman
    [75, 85, 89],  # Hassan
    [90, 87, 84]   # Kamran
]

# Task 4: Update Ali’s second grade to 95
grades[0][1] = 95  # Index 0 = Ali, index 1 = second grade

# Task 5: Add a new grade for Hassan in the fourth subject
grades[2].append(91)  # Index 2 = Hassan

# Task 6: Attempt to change Kamran’s roll number
try:
    roll_numbers[3] = 105  # Index 3 = Kamran
except TypeError as e:
    print("Error:", e)
    print("Explanation: Roll numbers are stored in a tuple, which is immutable. You cannot change individual elements.")

# Task 7: Print all students' names with updated grades
print("\nStudent Grades:")
print("---------------------------")
for i in range(len(students)):
    print(f"Name: {students[i]}")
    print(f"Roll Number: {roll_numbers[i]}")
    print(f"Grades: {grades[i]}")
    print("---------------------------")
```
