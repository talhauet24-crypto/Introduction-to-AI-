Find Total Impedance 

i. Add the three impedances.    
ii. Print real part (resistive) and imaginary part (reactive).       
iii. Print magnitude and phase angle (use abs() and cmath.phase()).




```python
import cmath


Z1 = complex(10, 5)     
Z2 = complex(4, -3)    
Z3 = complex(6, 12)     

```

```python
# Total impedance
Z_total = Z1 + Z2 + Z3

```

```python
# Real and imaginary parts
R_total = Z_total.real
X_total = Z_total.imag

```

```python
# Magnitude and phase
magnitude = abs(Z_total)
phase = cmath.phase(Z_total)   # radians


```

```python
print("Total Impedance:", Z_total)
print("Resistive Part (Real):", R_total)
print("Reactive Part (Imag):", X_total)
print("Magnitude:", magnitude)
print("Phase (radians):", phase)
```
