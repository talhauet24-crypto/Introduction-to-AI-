### 1️⃣ Load dataset and store in lists

```python
months = []
hours = []
electricity = []

with open(r"D:\Abdur Rehman\CEP\building_2.csv", "r") as f:
    header = f.readline().strip().split(",")

    month_i = header.index("Month")
    hour_i = header.index("Hour")
    elec_i = header.index("Equipment Electric Power [kWh]")

    for line in f:
        row = line.strip().split(",")

        months.append(int(float(row[month_i])))   # ✅ FIX
        hours.append(int(float(row[hour_i])))     # safer
        electricity.append(float(row[elec_i]))
```

### 2️⃣ Annual consumption

```python
annual_consumption = sum(electricity)
print("Annual Consumption (kWh):", round(annual_consumption, 2))
```

### 3️⃣ Monthly consumption

```python
monthly_consumption = {}

for m, e in zip(months, electricity):
    if m not in monthly_consumption:
        monthly_consumption[m] = 0
    monthly_consumption[m] += e

print("\nMonthly Consumption (kWh):")
for m in sorted(monthly_consumption):
    print("Month", m, ":", round(monthly_consumption[m], 2))
```

### 4️⃣ Daily consumption

```python
daily_consumption = []
daily_sum = 0

for i in range(len(electricity)):
    daily_sum += electricity[i]

    if (i + 1) % 24 == 0:
        daily_consumption.append(daily_sum)
        daily_sum = 0

print("\nTotal Days:", len(daily_consumption))
```

### 5️⃣ Seasonal consumption

```python
seasonal = {
    "Winter": 0,
    "Spring": 0,
    "Summer": 0,
    "Autumn": 0
}

for m, e in zip(months, electricity):
    if m in [12, 1, 2]:
        seasonal["Winter"] += e
    elif m in [3, 4, 5]:
        seasonal["Spring"] += e
    elif m in [6, 7, 8]:
        seasonal["Summer"] += e
    else:
        seasonal["Autumn"] += e

print("\nSeasonal Consumption (kWh):")
for s in seasonal:
    print(s, ":", round(seasonal[s], 2))
```

### 6️⃣ Weekday vs Weekend consumption

```python
weekday = 0
weekend = 0

for day, value in enumerate(daily_consumption):
    if day % 7 in [5, 6]:   # Saturday, Sunday
        weekend += value
    else:
        weekday += value

print("\nWeekday Consumption (kWh):", round(weekday, 2))
print("Weekend Consumption (kWh):", round(weekend, 2))
```

### 7️⃣ Visualization 

```python
daily = []
total = 0

for i in range(len(electricity)):
    total += electricity[i]
    if (i + 1) % 24 == 0:
        daily.append(total)
        total = 0

```

```python
print("\nDaily Consumption (first 10 days):")
for i in range(10):
    print("Day", i+1, ":", round(daily[i], 2), "kWh")
```
