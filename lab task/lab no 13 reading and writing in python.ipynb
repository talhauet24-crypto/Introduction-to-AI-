### Lab Practice Task:

Load Demand Data Recording    
An electrical feeder’s hourly load demand (kW) needs to be recorded for monitoring.    
Tasks:     
	Create a CSV file to store hourly load data     
	Write load values with time stamps    
	Organize each record on a separate line     


```python
import csv
from datetime import datetime

# Create CSV file and write data
with open("load_demand.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    
    # Header row
    writer.writerow(["Time", "Load Demand (kW)"])
    
    # Sample hourly load data
    load_data = [
        ("08:00", 120),
        ("09:00", 150),
        ("10:00", 180),
        ("11:00", 200),
        ("12:00", 220)
    ]
    
    # Write each record on a separate line
    for record in load_data:
        writer.writerow(record)

print("Load demand data recorded successfully.")
```

### Solar Energy Generation Logging
A solar PV system records daily energy generation (kWh).   
Tasks:   
	Create a file to store daily generation data    
	Write the header and initial records       
	Append new daily values without deleting old data    



```python
import csv

with open("solar_generation.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Energy_kWh"])
    writer.writerow(["2025-01-01", 45.6])
    writer.writerow(["2025-01-02", 48.2])
```

### Task: Organize and Process CSV Data
Read the CSV file line by line and:   
	Count the total number of recorded days    
	Display each record with a serial number   
The header must not be counted as a data record.   



```python
import csv

with open("solar_generation.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)

    count = 0
    for i, row in enumerate(reader, start=1):
        print(f"{i}. {row}")
        count += 1

print("Total Recorded Days:", count)
```

### Task: Write Organized Data to a New CSV File
Create a new file named organized_energy_data.csv.   
Write the processed and formatted energy records into this file in CSV format.    
The original file must remain unchanged    

```python
import csv

with open("solar_generation.csv", "r") as infile, \
     open("organized_energy_data.csv", "w", newline="") as outfile:

    reader = csv.reader(infile)
    writer = csv.writer(outfile)

    header = next(reader)
    writer.writerow(["S.No"] + header)

    for i, row in enumerate(reader, start=1):
        writer.writerow([i] + row)
```
