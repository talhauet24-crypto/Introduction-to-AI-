### **5)  Hospital Patient Monitoring System**
Patient vitals are recorded every 10 minutes:
{
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}
}
1)	Print the oxygen level.
2)	Add a new medication "antibiotic": "6PM".
3)	Update heart rate to 78.
4)	Get the medication schedule for "insulin" using .get() with default "Not Prescribed".
5)	Use .items() to print all key-value pairs.


```python
patient = {
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}
}
```

```python
print("Oxygen Level:", patient["oxygen_level"])
```

```python
patient["medications"]["antibiotic"] = "6PM"
print("Updated Medications:", patient["medications"])
```

```python
patient["heart_rate"] = 78
print("Updated Heart Rate:", patient["heart_rate"])
```

```python
insulin_schedule = patient["medications"].get("insulin", "Not Prescribed")
print("Insulin Schedule:", insulin_schedule)
```

```python
print("All Patient Info:")
for key, value in patient.items():
    print(f"{key}: {value}")
```
