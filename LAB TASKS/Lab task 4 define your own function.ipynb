#### 3.	 Solar Battery Charging:
A solar inverter monitors the battery’s state of charge (SOC).   
•	Write a function charge_from_solar(SOC) that:     
o	Prints "Charging" if SOC < 20%.         
o	Prints "Battery full, no action" if SOC > 90%.     
o	Prints "Battery charging, SOC = ---%" otherwise.    
•	Add docstrings and function annotations.     
•	Print the function’s docstring to check.     


```python
#lab task 03
def charge_from_solar(SOC: int)-> str:#innotation is used to tell the type of argument but its not compulsary
    '''this is docstring works like 
            multiple  line comment'''
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print(f"Battery charging, SOC = {SOC}%")
    print(charge_from_solar.__annotations__)

charge_from_solar(10)
charge_from_solar(95)
charge_from_solar(50)
```

#### 4.	Resistance, Voltage, and Current Lists:
You are building a program to analyze five resistor values in a circuit.       
•	Function 1: resistance():  return a list of 5 resistances.        
•	Function 2: voltage():  return voltages across each resistor.     
•	Function 3: current():  take both lists (Resistance & Voltage) and calculate currents.      
•	Display all lists nicely in the main program.      
•	Try two versions:      
o	Return as a tuple.       
o	Return as a dictionary with keys like "current_1".     


```python
#lab n0 4
def resistance():
    return [10, 20, 30, 40, 50]

def voltage_list():
    return [5, 10, 15, 20, 25]

def current_dictionary(R,V) -> dict:
    return {f"current_{i+1}": V[i]/R[i] for i in range(len(R))}
def current_touple(R,V):
    list1=[]
    for i in range(len(R)):
     list1.append(V[i]/R[i])
    return tuple(list1)  
R = resistance()
V = voltage_list()
I = current_dictionary(R, V)
I2 = current_touple(R,V)
print("Resistances:", R)
print("Voltages:", V)
print("Currents:", I)
print("Currents:", I2)
```

#### 7.	Series Resistance with *args:
You now want to make the series resistance function more flexible.      
•	Define series(*args) using tuple packing.      
•	Pass any number of resistances.      
•	Return the total resistance.        
•	Add docstrings and annotations.    
•	Print __annotations__.     



```python
#lab task 06 and 7
def series_args(*args: float) -> float:
    """Series resistance using *args"""
    return sum(args)

def parallel_args(*args: float) -> float:
    """Parallel resistance using *args"""
    return 1 / sum(1/r for r in args)

print("Series(*args):", series_args(10, 20, 30))
print("Parallel(*args):", parallel_args(10, 20, 30))
```

#### 10.	Solar Inverter Data (Dictionary):

```python
#lab task 10
def solar_data_kwargs(**kwargs) -> None:
    """Print solar inverter data from keyword args"""
    for k, v in kwargs.items():
        print(f"{k}: {v}")

solar_data_kwargs(SOC=60, solar_generation=2, price=60)
```
