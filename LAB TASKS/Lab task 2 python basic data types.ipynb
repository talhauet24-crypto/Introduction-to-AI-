### Task 1: Set up Variables 

```python
resistance=15
current=1.2
voltage=resistance*current
#old way
print('value of resistance is %d ohm'%resistance)
print('current through the resistor is%.2f ampere'%current)
print('the voltage across the resistor of %d ohm is %.2f'%(resistance,voltage))
```

### Task 2: Old-Style Formatting (%) 

```python
resistance=15
current=1.2
voltage=resistance*current
#format way 
print('the voltage:{}v across resistor of {}ohm with current:{}ampere'.format(voltage,resistance,current))
print("Voltage across {0} Ω resistor with {1:.2f} A current is {2:.2f} V".format(resistance, current, voltage))
```

### Task 3: .format() Method 

```python
print("Resistor: { } Ω, Current: {} A, Voltage: {} V".format(resistance, current, voltage)) 
print("Voltage across {0} Ω resistor with { current } A current is { voltage} V".format(resistance, current, voltage))
```

### Task 4: f-Strings 


```python
resistance = 10 
current = 2.5324 
voltage = resistance * current
print(f"Resistor: {resistance} Ω, Current: {current:.2f} A, Voltage: {voltage:.2f} V")
```

### 3. Task: 
You are given a CSV file named integers.csv that contains one number per row (some 
are integers, and some are decimal numbers). 
Write a Python program that: 

*i. Reads all numbers from the CSV file. 

*ii. Converts them into floating-point numbers.

*iii. Counts how many of them are whole numbers (integers). 

*iv. Prints the total count of integers.
 


```python
import csv

with open(r'C:\Users\Dell\Desktop\python collection\integers.csv') as f:
    reader = csv.reader(f)
    numbers = [float(row[0])+2 for row in reader]
    print(numbers)
    count = sum(num.is_integer() for num in numbers)
    print("Total integers:", count)
```
