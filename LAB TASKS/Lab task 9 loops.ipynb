## Task 1:
A lab keeps records of resistor values. Students need to print each resistance value for verification.
Print each resistance value from a list of resistors.
	Create a list, then use for loop to display each item.


```python
# List of resistor values
resistor_values = [100, 220, 470, 560, 1000]

# Print each resistor value
for resistance in resistor_values:
    print(resistance)
```

## Task 2:
Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.
Print the square of each current reading using formula I².
	Make a list of current values, loop and multiply each by itself.


```python
# List of current readings
current_readings = [0.5, 1.0, 1.5, 2.0, 2.5]

# Calculate the square of each current (I²) and print
for current in current_readings:
    print(current**2)
```

# While Loop

## Task 1:
During charging, the battery slowly increases its percentage. Students want to simulate this charging process in code.
Keep printing "Charging..." until battery percentage reaches 100%.
	Start with battery = 0, increase by 10 in each loop, stop when 100.




```python
# Starting battery percentage
battery = 0

# Simulate charging process
while battery < 100:
    print("Charging...")
    battery += 10
```

## Task 2:
Engineers use countdown timers for testing. Students must simulate a timer that runs before starting a test.
Countdown timer from 10 to 0, showing "Test Running".
	Start variable = 10, decrease each loop, stop at 0.


```python
# Starting countdown timer
timer = 10

# Countdown from 10 to 0
while timer >= 0:
    print("Test Running")
    timer -= 1
```

## Task 3:
A voltage monitoring system asks for user input repeatedly until the system is turned off.
Keep asking user for voltage until they enter 0.
	Use input() in while loop, stop when entered value is 0.


```python
# Keep asking for voltage input until 0 is entered
while True:
    voltage = float(input("Enter voltage (0 to stop): "))
    if voltage == 0:
        break
```

## Task 4:
Power in a system doubles when load increases. Students need to simulate this doubling until it reaches a limit.
Keep doubling the power until it exceeds 1000 W.
	Start power = 1, multiply by 2 each loop, stop when >1000.


```python
# Starting power
power = 1

# Keep doubling the power until it exceeds 1000 W
while power <= 1000:
    print("Power:", power, "W")
    power *= 2
```

## Task 5:
            A machine keeps running until its cycle count reaches a limit. Students must simulate this with a loop.
Start from cycle = 1 and keep printing "Cycle Running" until cycle reaches 5.


```python
# Starting cycle count
cycle = 1

# Simulate machine running until cycle reaches 5
while cycle <= 5:
    print("Cycle Running:", cycle)
    cycle += 1
```

## Task 6:
            A motor runs until its speed reaches 3000 RPM. Students must simulate this by starting from 0 and adding 500 each loop.
Keep increasing speed by 500 until it reaches 3000, printing the speed each time.

```python
# Starting speed
speed = 0

# Increase speed by 500 until it reaches 3000 RPM
while speed < 3000:
    print("Motor Speed:", speed, "RPM")
    speed += 500
```

```python
import csv

file_path = r"E:\Introduction to Ai programming\Assignments\Assignment 1.csv"

hours = []
consumptions = []

with open(file_path, "r") as f:
    data = csv.DictReader(f)
    for row in data:
        hours.append(int(row["hour"]))
        consumptions.append(float(row["consumption"]))

total_consumption = sum(consumptions)
avg_consumption = total_consumption / len(consumptions)
max_consumption = max(consumptions)
min_consumption = min(consumptions)

kwh_values = [c / 1000 for c in consumptions]

print("Energy Consumption Report\n")
print("Total:", total_consumption, "Wh")
print("Average:", round(avg_consumption, 2), "Wh")
print("Max:", max_consumption, "Wh")
print("Min:", min_consumption, "Wh")

print("\nHourly (kWh):")
for h, k in zip(hours, kwh_values):
    print("Hour", h, "->", round(k, 3), "kWh")
```

```python
import csv
file_path = r"E:\Introduction to Ai programming\Assignments\Assignment 2.csv"
days = []
hours = []
consumptions = []
with open(file_path, "r") as f:
    data = csv.DictReader(f)
    for row in data:
        days.append(int(row["Day"]))
        hours.append(int(row["Hour"]))
        consumptions.append(float(row["consumption(w)"]))
# Group by hour
hour_groups = {}
for h, c in zip(hours, consumptions):
    if h not in hour_groups:
        hour_groups[h] = []
    hour_groups[h].append(c)
print("\n--- Consumption Grouped by Hour ---\n")
for h in sorted(hour_groups):
    values = hour_groups[h]
    avg_h = sum(values) / len(values)
    peak_h = max(values)
    print(f"Hour {h}: Values = {values}, Average = {round(avg_h,2)}, Peak = {peak_h}")
# Group by day
day_groups = {}
for d, c in zip(days, consumptions):
    if d not in day_groups:
        day_groups[d] = []
    day_groups[d].append(c)
print("\n--- Consumption Grouped by Day ---\n")
for d in sorted(day_groups):
    values = day_groups[d]
    total_d = sum(values)
    avg_d = total_d / len(values)
    print(f"Day {d}: Values = {values}, Total = {total_d}, Average = {round(avg_d,2)}")
```

```python
import csv

price_file = r"E:\Introduction to Ai programming\Assignments\Assignment 3 price.csv"
load_file = r"E:\Introduction to Ai programming\Assignments\Assignment 3 load.csv"

prices = []
loads = []

with open(price_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        prices.append(float(row["Price (Rs/KWh)"]))

with open(load_file, "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        loads.append(float(row["consumption(w)"]))

def compute_bill(loads, prices):
    loads_kwh = [l / 1000 for l in loads]  # convert W → kWh
    hourly_bill = [l * p for l, p in zip(loads_kwh, prices)]
    total_bill = sum(hourly_bill)
    return hourly_bill, total_bill

hourly_bill, total_bill = compute_bill(loads, prices)

print("\nHourly Bill (Rs):")
for i, bill in enumerate(hourly_bill, start=1):
    print(f"Hour {i}: Rs {bill}")

print(f"\nTotal Bill: Rs {total_bill}")
```
