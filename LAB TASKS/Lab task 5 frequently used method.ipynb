### Lab Task:
1. #### enumerate: 
enumerate () adds numbers (index) to items in a list or other collection.    
i.	A list contains voltage readings from 5 measurement points. Use enumerate () to display the reading number with each voltage.      
ii.	A student records resistance values of resistors in a lab. Use enumerate () to print the resistor number and its value.    
Example:
values = ['A', 'B', 'C']
for index, val in enumerate(values):
    print(index, val)
Output:
0 A
1 B
2 C



```python
# Task 1:enumerate() 
voltages = [220, 230, 225, 235, 240]
print("Voltage Readings:")
for index, value in enumerate(voltages):
    print( index + 1, ":", value, "V")

resistances = [100, 220, 330, 470, 560]
print("\nResistor Values:")
for index, value in enumerate(resistances):
    print(index + 1, ":", value, "Ohms")
```

#### zip: 
zip () combines two or more lists element by element.     
i.	To lists, one list has current readings, and another has voltage readings. Use zip() to print the pair of current and voltage.      
ii.	Combine a list of component names with their ratings (resistor with ohms, capacitor with µF).        
Example:
names = ['a', 'b', 'c']   
nums = [1, 2, 3]   
for n, v in zip(name , nums):   
    print(n, v)   
Output:

```python
    # Task 2: zip() 
currents = [5, 10, 15, 20]
voltages = [110, 220, 330, 440]
print("Current and Voltage Pairs:")
pairs=list(zip(currents,voltages))
print(pairs)

components = ["Resistor", "Capacitor", "Inductor"]
ratings = ["100 Ohm", "10 µF", "1 H"]
print("\nComponents and Ratings:")
for name, rating in zip(components, ratings):
    print(name, ":", rating)
```

### map: 
 map() applies a function to every item in a list   
Note: lambda is a short way to write a function in one line. Example: lambda x: x*2 means a function that doubles x.    
i.	Convert a list of temperatures from Celsius to Fahrenheit using the formula F = (C × 9/5) + 32.    
ii.	Convert a list of voltages in volts to kilovolts using V_kV = V / 1000.     
Example:    
nums = [1, 2, 3]    
doubles = list(map(lambda x: x*2, nums))     
print(doubles)     
Output:      
[2, 4, 6]     


```python
# Task 3: map()
temps_c = [0, 25, 50, 100]
def c_to_f(x):
    return (x * 9/5) + 32
temps_f = list(map(c_to_f, temps_c))
print("Temperatures in Fahrenheit:", temps_f)
voltages_v = [1000, 2000, 3000]
def v_to_kv(v):
    return v / 1000
voltages_kv = list(map(v_to_kv, voltages_v))
print("Voltages in kV:", voltages_kv)
```

### filter()
filter() selects only the items that match a condition.   
i.	A list contains current readings. Use filter() to keep only readings greater than 10 A.    
ii.	A capacitor bank has values [10, 20, 50, 100]. Use filter() to select capacitors above 50 µF.    
Example:   
nums = [1, 2, 3, 4, 5]    
even = list(filter(lambda x: x % 2 == 0, nums))   
print(even)   
Output:    
[2, 4]   


```python
# Task 4: filter()
currents = [5, 12, 8, 15, 3]
def greater_than_10(x):
    return x > 10
filtered_currents = list(filter(greater_than_10, currents))
print("Currents greater than 10A:", filtered_currents)
capacitors = [10, 20, 50, 100]
def above_50(x):
    return x > 50
filtered_caps = list(filter(above_50, capacitors))
print("Capacitors above 50 µF:", filtered_caps)
```

### any: 
 any() returns True if at least one item in a list is True.    
i.	Check if any voltage reading in a list is above 220 V.    
ii.	Check if any motor is overloaded from a list of load status values.    
Example:   
values = [0, 0, 1, 0]    
print(any(values))    
Output:   
True   



```python
# Task 8: any()
voltages = [200, 210, 240, 280]
check_voltage = any(v > 220 for v in voltages)
print("Any voltage above 220V?", check_voltage)
motors = [False, False, True, False]
print("Any motor overloaded?", any(motors))
```

```python
# Task 9: all()
resistors = [150, 200, 10, 300]
check_resistance = all(r > 100 for r in resistors)
print("All resistors above 100 ohms?", check_resistance)
solar_panels = [True, True, True]
print("All solar panels active?", all(solar_panels))
```
