### 2.	Task: Pass/Fail Test Report (Engineering Labs)
i.	A file stores students’ lab test scores.     
ii.	Task: Read scores from file and apply conditions:   
a.	Pass if ≥ 50.    
b.	Fail if < 50.    


```python
#lab task 02
with open("score.txt","r") as f:
    scores = f.readlines()

results = []
for s in scores:
    marks = int(s.strip())
    if marks >= 50:
        results.append("Pass")
    else:
        results.append("Fail")

with open("results.txt", "w") as f:
    for r in results:
        f.write(r + "\n")
    print(results)
```

### 3.	Task: Current Rating Check (Electrical Machines)
i.	A file has current values (in Amperes).     
ii.	Task: Check for each line:     
a.	If current < 5A → Safe     
b.	If 5A ≤ current < 10A → Warning    
c.	If ≥ 10A → Overload    


```python
#labtask 03
with open("current.txt", "r") as f:
    currents = f.readlines()

results = []
for c in currents:
    current = float(c.strip())
    if current < 5:
        results.append("Safe")
    elif current < 10:
        results.append("Warning")
    else:
        results.append("Overload")

with open("current_results.txt", "w") as f:
    for r in results:
        f.write(r + "\n")
    print(results)
```

### 4.	Task: File Extension Validator (General Engineering Reports)
i.	A directory has multiple engineering reports.
ii.	Task: Write a program to scan all files.
a.	If file ends with .pdf → Valid Report
b.	If file ends with .docx → Needs Conversion
c.	Otherwise → Invalid Format
iii.	Count and display the numbers of each type.


```python
#labtask 04
import os

path = "reports"
valid = needs_conv = invalid = 0

for filename in os.listdir(path):
    if filename.endswith(".pdf"):
        print(filename, "→ Valid Report")
        valid += 1
    elif filename.endswith(".docx"):
        print(filename, "→ Needs Conversion")
        needs_conv += 1
    else:
        print(filename, "→ Invalid Format")
        invalid += 1

print("\nSummary:")
print("Valid Reports:", valid)
print("Needs Conversion:", needs_conv)
print("Invalid Format:", invalid)
```
