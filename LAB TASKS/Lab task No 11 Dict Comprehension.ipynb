### Task 1:    
Map Device Names to Their Power Ratings    
Given:    
devices = ["Fan", "Light", "AC", "Pump"]    
power = [60, 15, 1800, 750]     
Create a dict using comprehension:     
{"Fan": 60, "Light": 15, ...}    




```python
devices = ["Fan", "Light", "AC", "Pump"]
power = [60, 15, 1800, 750]

device_power = {devices[i]: power[i] for i in range(len(devices))}

print(device_power)
```

### Task 2:
Categorize Resistances as “Low” or “High”     
Resistance values (ohms):    
R = [5, 33, 47, 82, 10]     
Create a dict:     
Key = value, Value = "Low" if R < 40 else "High".     


```python
R = [5, 33, 47, 82, 10]

resistance_dict = {r: ("Low" if r < 40 else "High") for r in R}

print(resistance_dict)
```

### Task 3: 
Compute Current for Each Load    
Loads in watts:    
loads = [60, 100, 150, 200]     
Assume voltage = 200 V    
Dictionary comprehension to produce:     
{load: load/200} (I = P/V)     



```python
loads = [60, 100, 150, 200]
voltage = 200

current_dict = {load: load / voltage for load in loads}

print(current_dict)
```

### **Task 4:** 
Convert Dictionary Values From °C to °F     
Given:    
temps_c = {"motor": 40, "transformer": 60, "room": 28}     
Use dict comprehension to convert to Fahrenheit.    


```python
temps_c = {"motor": 40, "transformer": 60, "room": 28}

temps_f = {key: (value * 9/5) + 32 for key, value in temps_c.items()}

print(temps_f)

```

### **Task 5:** 
Create Dict of Components With Index Keys     
Given component names:    
components = ["Resistor", "Capacitor", "Inductor", "Diode"]    
Create {0:"Resistor", 1:"Capacitor", ...}    
Using dict comprehension + enumerate.   



```python
components = ["Resistor", "Capacitor", "Inductor", "Diode"]

component_dict = {i: name for i, name in enumerate(components)}

print(component_dict)
```

### **Task 6:**
Filter Only Devices With Power > 100W     
Given:         
device_power = {"Fan": 60, "Light": 15, "Heater": 1200, "Iron": 1000, "Router": 10}           
Use dict comprehension to keep only items where value > 100.             


```python
device_power = {"Fan": 60, "Light": 15, "Heater": 1200, "Iron": 1000, "Router": 10}

high_power_devices = {device: power for device, power in device_power.items() if power > 100}

print(high_power_devices)
```
