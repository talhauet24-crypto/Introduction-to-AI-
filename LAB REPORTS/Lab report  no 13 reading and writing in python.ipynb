### Task 1: Create and Write Energy Data to a CSV File
Create a file named energy_data.csv.    
Write the header and daily energy consumption records into the file using the CSV format:   
	Date    
	Energy consumed in kWh   
Each record must be written on a new line following the header.   


```python
from datetime import datetime

# Get current date and time
now = datetime.now()
print(now) 

# Format date as a string
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print(formatted)  

# Create a specific datetime
my_date = datetime(2025, 12, 25, 10, 30)
print(my_date)
```

### Task 2: Read and Display CSV File Data
Read the contents of energy_data.csv and display all energy consumption records in a clear and readable format.   
The data must be read from the file, not hardcoded.    


```python
import csv

# Open the CSV file in read mode
with open(r'') as file:

    reader = csv.reader(file)  # Create a CSV reader object

    # Read the header row
    header = next(reader)
    print(" | ".join(header))  # Display headers
    print("-" * 50)

    # Read and display each row
    for row in reader:
        print(" | ".join(row))
```

### Task 3: Append New Energy Records
Add new daily energy consumption records to energy_data.csv without removing existing data.   
Ensure:   
	The CSV format is preserved   
	New data is appended correctly   



```python
import csv

# New records to append (list of lists)
new_records = [
    ['2025-12-03', 'AC', '13.2'],
    ['2025-12-03', 'Heater', '8.7'],
    ['2025-12-03', 'Lights', '3.9']
]

# Open CSV file in append mode ('a')
with open('energy_data.csv', 'a', newline='') as file:
    writer = csv.writer(file)

    # Append each new record
    for record in new_records:
        writer.writerow(record)

print("New records appended successfully!")
```
